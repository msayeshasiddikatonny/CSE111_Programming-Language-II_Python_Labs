```python
class Smartphone:
  count = 0
  def __init__(self, name, year):
    self.name = name
    self.year = year
    self.feature = []
    Smartphone.count += 1
  def add_Features(self, *args):
    for x in args:
     self.feature.append(x)
  def printPhoneDetail(self):
    print(f"Model Name: {self.name}")
    print(f"Year of release: {self.year}")
    print(f"Features: ", end = "")
    print(self.feature[0], end = "")
    for i in range(1, len(self.feature)):
     print(", " + self.feature[i], end = "")
    print()

print("No.of Smartphone =", Smartphone.count)
s1 = Smartphone('Oneplus 9 Pro', 2021)
s1.add_Features('Snapdragon 888' , 'Wireless Charging', 'Hasselblad Camera')
s2 = Smartphone('Samsung Galaxy S21', 2020)
s2.add_Features('Exynos 2100' , 'Fast charging 25W')
s3 = Smartphone('MI 11 Ultra', 2021)
s3.add_Features('Fast charging 67W' , 'Snapdragon 888')
print("=========================")
s1.printPhoneDetail()
print("=========================")
s2.printPhoneDetail()
print("=========================")
s3.printPhoneDetail()
print("=========================")
print("No.of Smartphone =",Smartphone.count)
```

```python
class ECustomer:
    count=0
    def __init__(self,name):
      self.name=name
      self.items=[]
      self.costs=[]
      ECustomer.count+=1
    def setProductDetails(self,*infos):
      for info in infos:
       if isinstance(info,int)==True:
        self.costs.append(info)
       else:
        self.items.append(info)
    def printDetail(self):
      total=0
      for price in self.costs:
        total+=price
      total+=200
      if total<10000:
       new=float(total)
      elif 10000<total<50000:
       new=total+(0.05*total)
      elif total>50000:
       new=total+(0.1*total)
      print("Name: ",self.name)
      print("Total cost with delivery charge:",new)
# Do not change the following lines of code.
print("Total E-Customer:", ECustomer.count)
c1 = ECustomer("James")
c1.setProductDetails("TV",35000,"Air Cooler", 9000)
c2 = ECustomer("Mike")
c2.setProductDetails("Mobile",20000,"Headphone", 1200, "Fridge", 45000)
c3 = ECustomer("Sarah")
c3.setProductDetails("Headphone", 1200)
c4 = ECustomer("John")
c4.setProductDetails("AC",60000)
print("=========================")
c1.printDetail()
print("=========================")
c2.printDetail()
print("=========================")
c3.printDetail()
print("=========================")
c4.printDetail()
print("=========================")
print("Total E-Customer:", ECustomer.count)
```

```python
class Harry_Potter:
    book_count = 0

    def __init__(self, name, publish_year):
        self.name = name
        self.publish_year = publish_year
        Harry_Potter.book_count += 1

    def display_book_info(self):
        return 'Book---> ' + str(self.publish_year) + ": " + self.name

    def set_name(self, name):
        self.name = name

    def set_year(self, year):
        self.publish_year = year
 
# Do not change the following lines of code.
print("Book Count:", Harry_Potter.book_count)
print("=======================")
b1 = Harry_Potter("Harry Potter and the Philosopher's Stone.",1997)
print(b1.display_book_info())
b2 = Harry_Potter("Harry Potter and the Chamber of Secrets.", 0)
b2.set_year(1998)
print(b2.display_book_info())
b3 = Harry_Potter("Harry Potter and the Prisoner of Azkaban.",1999)
print(b3.display_book_info())
b4 = Harry_Potter("Default", 0)
b4.set_name("Harry Potter and the Goblet of Fire.")
b4.set_year(2000)
print(b4.display_book_info())
print("=======================")
print("Book Count:", Harry_Potter.book_count)
```
