```python
class Employee:
    total_employee = 0
    
    def __init__(self, name, work_hour):
        self.name = name
        self.work_hour = work_hour
    
    def __str__(self):
        s =  "Department: "+self.name+", Work hours: "+str(self.work_hour)
        return s

# Write your codes here.
class FullTimeEmployee(Employee):
    def __init__(self, name, work_hour):
        super().__init__(name, work_hour)
        self.employees = []
        print(f'Full time employees in {name} have to work {work_hour} hours')

    def addFullTimeEmployee(self, *employee_details):
        for i in range(0, len(employee_details), 2):
            self.employees.append((employee_details[i], employee_details[i + 1]))
        Employee.total_employee += len(employee_details) // 2
    
    def __str__(self):
        s = super().__str__()
        s += "\nTotal Employee(s): " + str(len(self.employees))
        s += "\nEmployee details:"
        for employee in self.employees:
            s += "\nName: " + employee[0] + ", Work hours remaining: " + str(self.work_hour - employee[1])
        
        return s

class PartTimeEmployee(Employee):
    def __init__(self, name, work_hour):
        super().__init__(name, work_hour)
        self.employees = []
        print(f'Part time employees in {name} have to work {work_hour} hours')

    def addPartTimeEmployee(self, *employee_details):
        for i in range(0, len(employee_details), 2):
            self.employees.append((employee_details[i], employee_details[i + 1]))
        Employee.total_employee += len(employee_details) // 2
    
    def __str__(self):
        s = super().__str__()
        s += "\nTotal Employee(s): " + str(len(self.employees))
        s += "\nEmployee details:"
        for employee in self.employees:
            s += "\nName: " + employee[0] + ", Work hours remaining: " + str(self.work_hour - employee[1])
        
        return s

# Do not change the following lines of code.
p1 = FullTimeEmployee("Finance", 40 )
print("=================================")
p1.addFullTimeEmployee("Bob", 12,  "Carol", 18, "Mike", 15)
print("=================================")
print(p1)
print("=================================")
p2 = PartTimeEmployee("Others", 25)
print("=================================")
p2.addPartTimeEmployee("David", 12, "Simon", 18)
print("=================================")
print(p2)
print("=================================")
print("Total Employee ", Employee.total_employee)
```
